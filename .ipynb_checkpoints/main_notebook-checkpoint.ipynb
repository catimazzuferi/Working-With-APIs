
# DISNEY

```python
import pandas as pd
import requests
from pandas.io.json import json_normalize

response = requests.get('https://api.disneyapi.dev/character')
data = response.json()

if response.status_code == 200:
    
    data = pd.json_normalize(data['data'])
    data.head()
else:
    print(f"Error: {response.status_code}")
```
